## OUTLIER DETECTION

```python
#  - Flag txns above 99th percentile
q99 = df['amt'].quantile(0.99)
#creating a new column where encoding -> where 1 indicates high value transaction outlier and 0 indicates normal transaction
df['amt_outlier'] = (df['amt'] > q99).astype(int)
#99% of all transactions are below this value which is $545.99 so only 1 % of transactions are higher than this
# might pass this encoding into a model to predict fraud (behaviour at large spending values)
print(f"99th percentile of amt = {q99:.2f}; flagged amt_outlier")
```

# Why this matters
## Baseline “dumb” classifier

### A model that always predicts “not fraud” would be 99.42% accurate. That sounds better than it actually is since it never catches a single fraud. It would miss those cases of actual fraud causing loss. <br>


### Misleading accuracy
In imbalanced settings, accuracy is almost worthless. You need metrics that emphasize the minority (fraud) class. <br>


Choosing the right evaluation metrics <br>
Precision	Of all transactions flagged as fraud, what fraction truly were fraud? <br>
Recall	Of all actual frauds, what fraction did you catch? <br>
F1-Score	Harmonic mean of precision & recall; balances false positives & negatives. <br>
ROC-AUC	Model’s ability to rank fraud vs non-fraud across all thresholds. <br>
PR-AUC	Area under the Precision-Recall curve; more informative on imbalanced data. <br>

When training will have to split on the fraud column so that train/val/test all have the same fraud rate ~0.58%
This prevents model from seeing different fraud rates in each split. <br>

## Algorithmic Approaches <br>
### Class weights / Cost Sensitive Learning <br>
### -Tell the model that mistakes on fraud cost more than on normal transactions <br>
### -Anomaly detection which will test fraud as anomalies since train model will learn that normal transactions are not frauds (autoencoders) <br>

Note:
When fraud is rare, PR-AUC is often more meaningful than ROC-AUC, because PR-AUC focuses solely on performance for the positive (fraud) class.
